```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Jornada Python - Hashtag/Aula 3 - Python IA/clientes.csv')
df
```

```python
df.info()
```

```python
#coluna score_credito:
#Good -> bom; Standard -> ok; Poor -> ruim;
```

```python
df['profissao'].value_counts()
```

```python
#importando o label encoder:
#usando a biblioteca de IA -> scikit-learn
#sklearn
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()

#codificando a coluna profissão:
df['profissao'] = codificador.fit_transform(df['profissao'])
#transforma os valores da coluna profissão em numerico

df['mix_credito'] = codificador.fit_transform(df['mix_credito'])
df['comportamento_pagamento'] = codificador.fit_transform(df['comportamento_pagamento'])

df
```

```python
df.info()
```

```python
#para a IA aprender, precisamos fazer duas divisões na base de dados

#1a: (Y) Quem eu quero prever e Quem eu quero usar na previsão;
# (X) o que vou usar pra fazer a previsão

y = df['score_credito']
x = df.drop(columns=['score_credito', 'id_cliente'])

#2a: separar os dados de X e Y em dados de 'treino' e dados de 'teste'
#x_treino e x_teste; y_treino e y_teste;
```

```python
#treino e teste:
from sklearn.model_selection import train_test_split #separa os dados em  treino e teste

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y)
#tem que ser nessa ordem (treino, teste, treino, teste)
#test_size=percentual -> parametro dado na função para especificar uma pocentagem para treino
#ex: test_size=0.3
```

```python
#criando a inteligencia artificial:
#importando a inteligencia artificial:
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#cria a intligencia artificial:
#modelo de arvore de decisão
modelo_arvoredecisao = RandomForestClassifier()
```

```python
#modelo KNN -> vizinhos proximos (Nearest Neighbors)
modelo_knn = KNeighborsClassifier()
```

```python
#treinando a IA:
modelo_arvoredecisao.fit(x_treino,y_treino)
modelo_knn.fit(x_treino,y_treino)
```

```python
#testar o modelo:
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

#verificando a acuracia das previsões:
from sklearn.metrics import accuracy_score

print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_knn)) #compara o resultado da IA com o Gabarito (y_teste)
```

```python
#o melhor modelo de previsão é o de arvore de decisão
```
